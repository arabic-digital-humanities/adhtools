```python
doc_labels = document_topics.idxmax(axis=1).map(lambda i: topics.loc[int(i), 'label'])
```

```python
data['label'] = doc_labels
```

```python
import seaborn as sns
topic_names = topics['label']
color_list = sns.color_palette("muted", n_colors=10) + sns.color_palette("dark", n_colors=10)
color_dict = {label: color_list[i] for i,label in enumerate(topic_names)}
color_dict['noise'] = (0.9, 0.9, 0.9)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 15))
c = [color_dict[l] for l in data.label]
for label in data.label.unique():
    c = color_dict[label]
    data_sub = data[data.label==label]
    plt.scatter(data_sub['x'], data_sub['y'], c=[c], label=label)
plt.legend()
plt.show()
```
