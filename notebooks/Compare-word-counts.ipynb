```python
%matplotlib inline
import matplotlib.pyplot as plt
import adhtools.utils
import pandas as pd
import os
import numpy as np
```

```python
import glob

files_light10 = glob.glob('/media/sf_VBox_Shared/Arabic/Fiqh/2018-11-06-Fiqh-light10/*.xml')
fnames_light10 = [os.path.basename(fn) for fn in files_light10]
corpus_light10 = adhtools.utils.corpus_wordlist(files_light10, analyzer=False, field='proposed_root')
lengths_light10 = {fnames_light10[i]: len(s) for i, s in enumerate(corpus_light10)}
lengths_light10 = {k[:-4]: lengths_light10[k] for k in lengths_light10}
```

```python
files_alkhalil = glob.glob('/media/sf_VBox_Shared/Arabic/Fiqh/2018-09-18-Al-Khalil/*.xml')
fnames_alkhalil = [os.path.basename(fn) for fn in files_alkhalil]
corpus_alkhalil = adhtools.utils.corpus_wordlist(files_alkhalil, analyzer=True, field='proposed_root')
lengths_alkhalil = {fnames_alkhalil[i]: len(s) for i, s in enumerate(corpus_alkhalil)}
```

```python
lengths_alkhalil = {k[:-4]: lengths_alkhalil[k] for k in lengths_alkhalil}
```

```python
files_txt = glob.glob('/media/sf_VBox_Shared/Arabic/fiqh_corpus/txt/*.txt')
lengths_txt = {}
for fn in files_txt:
    with open(fn) as f:
        s = f.read()
    l = len(s.split(' '))
    lengths_txt[os.path.basename(fn)] = l
lengths_txt = {k[:-4]: lengths_txt[k] for k in lengths_txt}
```

```python
df = pd.DataFrame({'light10': lengths_light10, 'alkhalil': lengths_alkhalil, 'txt': lengths_txt})
```

```python
df.plot(kind='barh', figsize=(20, 30))
```
